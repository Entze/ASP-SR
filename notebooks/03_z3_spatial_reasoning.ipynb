```python
import math

import z3
```

```python
radius1 = z3.RealVal(1.0)
radius2 = z3.RealVal(2.0)
```

```python
solver = z3.Solver()
solver.add(math.pi * radius1**2 < math.pi * radius2**2)
solver
```

```python
solver.check()
```

```python
x_1 = z3.RealVal(0.0)
y_1 = z3.RealVal(0.0)
r_1 = z3.RealVal(1.0)

x_2 = z3.RealVal(1.0)
y_2 = z3.RealVal(0.0)
r_2 = z3.RealVal(5.0)

```

```python
d = z3.Sqrt((x_1 - x_2)**2 + (y_1 - y_2)**2)
```

```python
solver = z3.Solver()
```

```python
solver.add(r_2 > (d + r_1))
solver
```

```python
solver.check()
```
