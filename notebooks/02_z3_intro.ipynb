```python
import z3
```

```python
Object = z3.DeclareSort('Object')
Object
```

```python
Human = z3.Function('Human', Object, z3.BoolSort())
Human
```

```python
Mortal = z3.Function('Mortal', Object, z3.BoolSort())
Mortal
```

```python
socrates = z3.Const('socrates', Object)
socrates
```

```python
x = z3.Const('x', Object)
x
```

```python
axioms = [z3.ForAll([x], z3.Implies(Human(x), Mortal(x))), Human(socrates)]
axioms
```

```python
solver = z3.Solver()
```

```python
solver.add(axioms)
print(solver)
sat = solver.check()
print(sat)
print(solver.model())
```

```python
unsat = solver.check(z3.Not(Mortal(socrates)))
print(unsat)
print(solver.unsat_core())
```

```python
print(sat, sat == z3.sat, sat == z3.unsat)
print(unsat, unsat == z3.sat, unsat == z3.unsat)
```
